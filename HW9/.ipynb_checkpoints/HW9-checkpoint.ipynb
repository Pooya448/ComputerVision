```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

Run this cell for plotting.

```python
def plotter(img_list, r, w, gray, wr, hr, fig_name = None):
    '''
    Plots images' list with its' caption and saves result image if you want.

    Parameters:
        img_list (list): The list of tuples of image and its' caption.
        r (int): The number of row(s).
        w (int): The number of colunm(s).
        gray (bool): The flag for plotting images in grayscale mode.
        wr (int): The width of one figure.
        hr (int): The height of one figure.
        fig_name (str): The name of the image of the plot. if not set this parameter the plot doesn't save.
    '''
    
    plt.rcParams['figure.figsize'] = (wr, hr)
    for i in range(len(img_list)):
        plt.subplot(r, w, i + 1)
        if img_list[i][2] == 'img':
            if gray:
                plt.imshow(img_list[i][0], cmap = 'gray')
            else:
                plt.imshow(img_list[i][0])
            plt.xticks([])
            plt.yticks([])
        elif img_list[i][2] == 'hist':
            plt.bar(np.arange(len(img_list[i][0])), img_list[i][0], color = 'c')
        else:
            raise Exception("Only image or histogram. Use third parameter of tuples in img_list and set it to img or hist.")
        plt.title(img_list[i][1])
    if fig_name is not None:
        plt.savefig(fig_name + '.png')
    plt.show()
```

Implement this function for Augmented reality.

```python
def AR(background, image):
    '''
    Adds the input image to the background image properly.
    
    Parameters:
        background (numpy.ndarray) : background image
        image (numpy.ndarray): input image
    
    Returns:
        numpy.ndarray: The result image.
    '''
    result = background.copy()
    
    src_points = np.float32([
        [0, 0],
        [0, 1799],
        [1199, 1799],
        [1199, 0]
    ])
    dst_points = np.float32([
        [148, 105],
        [143, 225],
        [236, 228],
        [237, 105]
    ])
    
    h, status = cv2.findHomography(src_points, dst_points)

    image_perspective = cv2.warpPerspective(image, h, (background.shape[1], background.shape[0]))

    for i in range(background.shape[0]):
        for j in range(background.shape[1]):
            if image_perspective[i, j].all() > 0:
                result[i, j] = image_perspective[i, j]    
    
    return result
```

# PART 3A

Here is for your helper functions (optional).

```python
 def compute_histogram(image):
        
    histogram = np.zeros((256), np.int)
    np.add.at(histogram, image, 1)
    return histogram
```

Implement this function to get the global otsu algorithm output.

```python
def global_otsu(image):
    '''
    Applys global otsu on the input image.
    
    Parameters:
        image (numpy.ndarray): The input image.
    
    Returns:
        numpy.ndarray: The result panorama image.
    '''
    
    sigma_optim = 100000000
    threshold = 0
    
    histogram = compute_histogram(image)
    
    total_count = np.sum(histogram)

    for t in range(1, 255):
        P1 = histogram[:t]
        P2 = histogram[t:]
        
        Q1 = np.sum(P1)
        Q2 = np.sum(P2)
        
        W1 = Q1 / total_count
        W2 = Q2 / total_count
        
        V1 = np.var(P1)
        V2 = np.var(P2)
        
        Sigma = W1 * V1 + W2 * V2
        
        if Sigma < sigma_optim:
            sigma_optim = Sigma
            threshold = t
    
    image = np.where(image < threshold, 0, 255)
        
    
    return image
```

Here is for your helper functions (optional).

Implement this function to get local otsu algorithm output.

```python
def local_otsu(image):
    '''
    Applys local otsu on the input image.
    
    Parameters:
        image (numpy.ndarray): The input image.
    
    Returns:
        numpy.ndarray: The result panorama image.
    '''
    
    result = np.zeros_like(image)
    rows, cols = image.shape
    
    x_mid = rows // 2
    y_mid = cols // 2

    result[:x_mid, :y_mid] = global_otsu(image[:x_mid, :y_mid])
    result[:x_mid, y_mid:] = global_otsu(image[:x_mid, y_mid:])
    result[x_mid:, :y_mid] = global_otsu(image[x_mid:, :y_mid])
    result[x_mid:, y_mid:] = global_otsu(image[x_mid:, y_mid:])
    
    return result
```

```python
def adaptive_th(image):
    '''
    Applys adaptive threshold on the input image.
    
    Parameters:
        image (numpy.ndarray): The input image.
    
    Returns:
        numpy.ndarray: The result panorama image.
    '''
    
    result = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 19, 8)
    
    
    return result
```

```python
def ncc(image1, coords1, image2, coords2, win_size):
    
    offset = win_size // 2
    
    correspondence = []
    hist = []
    for point1 in coords1:
        score = -1
        most_scored = None
        for point2 in coords2:     
            slice1 = image1[point1[0] - offset:point1[0] + offset + 1, point1[1] - offset:point1[1] + offset + 1, :]
            slice2 = image2[point2[0] - offset:point2[0] + offset + 1, point2[1] - offset:point2[1] + offset + 1, :]

            res = cv2.matchTemplate(slice1, slice2, cv2.TM_CCORR_NORMED)
            hist.append(res)
            if res > score and res > 0.85:
                score = res
                most_scored = point2
                
        correspondence.append(most_scored)
    arr = np.array(hist)
    print(np.max(arr))
    return coords1, correspondence

def euclidean_distance(P1, P2):
    return np.sqrt(np.square(P1[0] - P2[0]) + np.square(P1[1] - P2[1]))
```

```python
def find_match(image1, image2):
    '''
    Finds match points between two input images.
    
    Parameters:
        image1 (numpy.ndarray): input image.
        image2 (numpy.ndarray): second input image.
    
    Returns:
        numpy.ndarray: The result image.
    '''
    
    distance_threshold = 15
    dist_offset = distance_threshold // 2
    
    discriptor_size = 19
    disc_offset = discriptor_size // 2
    
    kernel = np.ones((5,5), np.uint8) 
    
    gray1 = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(image2,cv2.COLOR_BGR2GRAY)
    
    corners1 = cv2.cornerHarris(gray1, 5, 3, 0.05)
    corners2 = cv2.cornerHarris(gray2, 5, 3, 0.05)
    
    dilated1 = cv2.dilate(corners1, None) 
    dilated2 = cv2.dilate(corners2, None) 
    
    points1 = dilated1 > 0.05 * dilated1.max()
    points2 = dilated2 > 0.05 * dilated2.max()
    
    rows, cols = points1.shape
    
    for i in range(rows):
        for j in range(cols):
            if points1[i, j]:
                if i + disc_offset >= rows or j + disc_offset >= cols or i - disc_offset < 0 or j - disc_offset < 0:
                    points1[i, j] = False
                else:
                    points1[i - dist_offset:i + dist_offset + 1, j - dist_offset:j + dist_offset + 1] = False
                    points1[i, j] = True
            if points2[i, j]:  
                if i + disc_offset >= rows or j + disc_offset >= cols or i - disc_offset < 0 or j - disc_offset < 0:
                    points2[i, j] = False
                else:
                    points2[i - dist_offset:i + dist_offset + 1, j - dist_offset:j + dist_offset + 1] = False
                    points2[i, j] = True
    
    
#     for i in range(points2.shape[0]):
#         for j in range(points2.shape[1]):
#             if points2[i, j]:
#                 if i + disc_offset >= points2.shape[0] or j + disc_offset >= points2.shape[1] or i - disc_offset < 0 or j - disc_offset < 0:
#                     points2[i, j] = False
#                 else:
#                     points2[i - dist_offset:i + dist_offset + 1, j - dist_offset:j + dist_offset + 1] = False
#                     points2[i, j] = True
    
    image1[points1] = [0,0,255]
    image2[points2] = [0,0,255]
    
    rows1, cols1 = np.where(points1)
    coords1 = list(zip(rows1, cols1))
    
    rows2, cols2 = np.where(points2)
    coords2 = list(zip(rows2, cols2))
    
    print(len(coords1))
            
    print(len(coords2))
    
    result1, result2 = ncc(image1, coords1, image2, coords2, discriptor_size)

    combined = np.concatenate([image1, image2], axis = -2)
    
    for i in range(len(result1)):
        r = np.random.randint(0, 255)
        g = np.random.randint(0, 255)
        b = np.random.randint(0, 255)
        
        src = (result1[i][1], result1[i][0])
        
        dst_y = result2[i][0]
        dst_x = result2[i][1] + image2.shape[1]
        dst = (dst_x, dst_y)
        cv2.line(combined, src, dst, (r, g, b), 2)
    
    return combined
```
